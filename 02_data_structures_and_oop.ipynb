## topic 3 - SETS 

A set is an unordered collection of unique elements.
It automatically removes duplicate values.
Sets are useful for storing distinct items.


```python
# creating a set with duplicate values
values = {1, 2, 3, 3, 4}

print(values)
```

## OOP BASICS

Object oriented programming is based on classes and objects 
A classs is a blueprint , and an object is an instance of class
it helps in organizing code in a structured way


```python
# defining a simple class

class student:
    def __init__(self,name,marks):
        self.name = name
        self.marks = marks 

    def display(self):
        print(self.name , self.marks)


# creating an object 
s1 = student('Shivam' , 87)
s1.display()



        
```
