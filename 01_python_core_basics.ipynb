Topic 1 - Variabe and Data types in python 

Variable is uded to store data .
Python is dynamically typed language , so we do not need to declare the data types.
data types already tells us what kind of value a variable is holding.
common data types - int , float , string , boolean 




```python
# storing basic information
age = 21          # integer
name = "Shivam"   # string
is_student = True # boolean

print(age)
print(name)
print(is_student)
```

Checking data types in python 

```python
# checking data types of variables
print(type(age))
print(type(name))
print(type(is_student))
```

Practice (Question + Solution)

```python
# Store your name, age, and whether you are a student.
# Print their values and data types.

# solution 
name = 'Shivam'
age = 20
student_status = 'True'

print(name , type(name))
print(age , type(age))
print(student_status , type(student_status))
print(f"My name is {name} and my age is {age} and my student status is {student_status} ")
```

**Takeaway:**  
Variables store data, and Python automatically assigns a data type based on the value.
