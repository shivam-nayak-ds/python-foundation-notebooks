## File Handling

File handling is used to read data from and write data to files.
Python provides built-in functions to work with text files.
It is commonly used for storing and processing data.


```python
# writing data to a file
with open("sample.txt", "w") as file:
    file.write("Learning Python file handling")

# reading data from the file
with open("sample.txt", "r") as file:
    print(file.read())
```

## EXCEPTION HANDLING

Exception handling is used to handle runtime errors.
It prevents the program from crashing unexpectedly.
The try-except block is commonly used for error handling.


```python
try:
    num = int(input("Enter a number: "))
    print(10 / num)
except ZeroDivisionError:
    print("Cannot divide by zero")
except ValueError:
    print("Invalid input")
```

## MODULES & PACKAGES

A module is a file that contains Python code.
It helps in organizing and reusing code.
Python provides many built-in modules.


```python
# using a built-in module
import math

print(math.sqrt(16))
```
